```python
from sklearn.impute import SimpleImputer
# Initialize the imputer
imputer = SimpleImputer(strategy='most_frequent')

# Apply the imputer to categorical columns
categorical_columns = ['refdocname']
df[categorical_columns] = imputer.fit_transform(df[categorical_columns])
```

```python
# drop the null columns
drop_columns=df.dropna(axis=1,inplace=True)
```

```python
groupby_columns=df.groupby('refdocname')['collection'].sum().reset_index()
```

```python
# Sort the DataFrame by 'collection' in descending order and get the top 5
top_5_doctors = groupby_columns.sort_values(by='collection', ascending=False).head(5)

# Display the result
print(top_5_doctors)
```

```python

# Set the size of the plot
plt.figure(figsize=(10, 6))  # Adjust width and height as needed

# Create the bar plot
sns.barplot(data=top_5_doctors, x='refdocname', y='collection', color='blue')

# Add title and labels
plt.title('Top 5 Doctors by Collection')
plt.xlabel('Doctor Name')
plt.ylabel('Total Collection')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)  # Rotate labels for better readability

# Show the plot
plt.tight_layout()  # Ensures labels fit within the figure
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df[['refdocname']]
y=df['collection']
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
x_encoded = encoder.fit_transform(df[['refdocname']])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.8,random_state=42)
```

```python
x_test
```

```python
y_test
```

```python
# initaling Model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(x_test)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```

```python
#step6: visulaize the results
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='r', linestyle='--')  # 45-degree line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Plot")
plt.show()
```
